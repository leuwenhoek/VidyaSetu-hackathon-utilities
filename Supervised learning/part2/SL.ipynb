```python
import pandas as pd
import os
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
DATA_PATH = os.path.join(os.path.expanduser("~"), "ayush", "hackathon", "Zonals", 
                        "VidyaSetu-hackathon-utilities", "Supervised learning", "Data", "cleaned_insurance.csv")

df = pd.read_csv(DATA_PATH)
df
```

## TRAINING and TESTING data split

```python
X = df.drop("charges",axis=1)
Y = df["charges"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
X_train
```

## Linear regression model Training

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,Y_train)
```

## TESTING and PREDICTION

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
import matplotlib.pyplot as plt
import numpy as np

# y_test & y_pred must be arrays
y_test = np.array(Y_test)
y_pred = np.array(X_test)

plt.figure(figsize=(8,6))

# Scatter plot
plt.scatter(Y_test, Y_pred)

# Perfect prediction line
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val])

plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.title("Actual vs Predicted (Regression Model)")
plt.grid(True)
plt.show()
```

## EVALUATING Model

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test,Y_pred)
r2
```
