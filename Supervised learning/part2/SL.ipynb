```python
import pandas as pd
import os
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
DATA_PATH = os.path.join(os.path.expanduser("~"), "ayush", "hackathon", "Zonals", 
                        "VidyaSetu-hackathon-utilities", "Supervised learning", "Data", "cleaned_insurance.csv")

df = pd.read_csv(DATA_PATH)
df
```

## TRAINING and TESTING data split

```python
X = df.drop("charges",axis=1)
Y = df["charges"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
X_train
```

## Linear regression model Training

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,Y_train)
```

## TESTING and PREDICTION

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
Y_test
```

## EVALUATING Model

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test,Y_pred)
r2
```

```python
n = X_test.shape[0]         #Rows
p = X_test.shape[1]         #Coluns

adjusted_r2 = 1 - ((1 - r2)*(n-1)/(n-p-1))          #fixed formula
adjusted_r2
```
