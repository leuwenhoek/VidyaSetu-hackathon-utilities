# Performing EDA

```python
df['sex'] = df['sex'].map({
    "female" : 0,
    "male" : 1,
})
df
```

```python
df['smoker'] = df['smoker'].map({
    "yes" : 1,
    "no" : 0
})
df
```

```python
df.rename(columns={"smoker":"is_smoker","sex":"is_male"},inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

df = pd.get_dummies(df,drop_first=True)
df
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
df['age'] = scale.fit_transform(df[['age']])
df
```

```python
df['bmi'] = scale.fit_transform(df[["bmi"]])
df
```

```python
cols = ["region_northwest","region_southeast","region_southwest"]

for col in cols:
    df[col] = df[col].map({
        False : 0,
        True : 1
    })

df
```

# co-relation

```python
cols = ['age', 'sex', 'bmi', 'children', 'smoker', 'charges',
       'region_northwest', 'region_southeast', 'region_southwest']
df_corr: pd.DataFrame
for col in cols:
    df_corr = abs((df.corr()[col]*100).round(2))

df_corr.sort_values(ascending=False)
```

# Model Training

## Linear regression

```python
x = df.drop("charges",axis=1)
y = df["charges"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
X_train = scale.fit_transform(X_train)
X_test = scale.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
li_pred = model.predict(X_test)
```

## Random forest

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=42)

random_forest_model = regressor.fit(X_train, y_train)
randomforest_y_pred = regressor.predict(X_test)
```

# Model evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_error,r2_score,root_mean_squared_error

mae = mean_absolute_error(y_test,li_pred)
mbe = mean_absolute_error(y_test,li_pred)
r2 = r2_score(y_test,li_pred)
rmse = root_mean_squared_error(y_test,li_pred)

print("linear regression MSE : ",mae)
print("linear regression MBE : ",mbe)
print("linear regression R2 : ",r2)
print("linear regression RMSE : ",rmse,"\n")

mae = mean_absolute_error(y_test,randomforest_y_pred)
mbe = mean_absolute_error(y_test,randomforest_y_pred)
r2 = r2_score(y_test,randomforest_y_pred)
rmse = root_mean_squared_error(y_test,randomforest_y_pred)

print("Random forest MSE : ",mae)
print("Random forest MBE : ",mbe)
print("Random forest R2 : ",r2)
print("Random forest RMSE : ",rmse)
```

# Saving the model

```python
import joblib

joblib.dump(random_forest_model,os.path.join(os.getcwd(),"model.pkl"))
```
