# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Basic layout to make a graph

```python
x = np.arange(1,11)
y = np.arange(1,11)
plt.title("Graph")
plt.xlabel("Nums 1")
plt.ylabel("Nums 2")
plt.plot(x,y)
```

## Subplots in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x*5,y*4)
plt.subplot(2,2,2)
plt.plot(x/2,y/3)
plt.subplot(2,2,3)
plt.plot(x//3,y//2)
plt.subplot(2,2,4)
plt.plot(x**3,y**2)
```
